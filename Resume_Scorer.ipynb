```python
import pdfplumber

def extract_text_from_pdf(pdf_path):
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text() + "\n"
    return text.strip()
```

```python
# Example usage
resume_text = extract_text_from_pdf("Tech_6.pdf")
print(resume_text[:5000])  # Print first 500 characters
```

```python
# Using basic regex to extract the details
import spacy
import re

# Load SpaCy NLP model
nlp = spacy.load("en_core_web_sm")

def extract_entities(text):
    doc = nlp(text)
    name = None
    email = None
    phone = None

    # Extract name (first entity)
    for ent in doc.ents:
        if ent.label_ == "PERSON":
            name = ent.text
            break

    # Extract email
    email_match = re.search(r"[a-zA-Z0-9+_.-]+@[a-zA-Z0-9.-]+\.[a-zA-Z]+", text)
    if email_match:
        email = email_match.group(0)

    # Extract phone number (basic pattern for 10-digit numbers)
    phone_match = re.search(r"\b\d{10}\b", text)
    if phone_match:
        phone = phone_match.group(0)

    return {"name": name, "email": email, "phone": phone}
```

```python
# Uses transformers model based on Name Entity Recognition to extract
import re
from transformers import pipeline

# Load NER model
ner_pipeline = pipeline("ner", model="dslim/bert-base-NER", aggregation_strategy="simple")

def extract_entities(text):
    ner_results = ner_pipeline(text)
    name = None
    email = None
    phone = None

    # Extract name from NER results
    for entity in ner_results:
        if entity["entity_group"] == "PER":
            name = entity["word"]
            break

    # Extract email using regex
    email_match = re.search(r"[a-zA-Z0-9+_.-]+@[a-zA-Z0-9.-]+\.[a-zA-Z]+", text)
    if email_match:
        email = email_match.group(0)

    # Extract phone number (detects numbers with country codes too)
    phone_match = re.search(r"(\+\d{1,3}[-.\s]?)?\(?\d{3}\)?[-.\s]?\d{3}[-.\s]?\d{4}", text)
    if phone_match:
        phone = phone_match.group(0)

    return {"name": name, "email": email, "phone": phone}
```

```python
entities = extract_entities(resume_text)
print(entities)
```

```python
SKILLS_DB = {"Python", "Machine Learning", "Deep Learning", "AI", "NLP", "Data Science", "TensorFlow", "PyTorch", 
             "SQL", "Java", "C++", "Keras", "Flask", "Django", "AWS", "Docker", "Kubernetes", "Spark"}

def extract_skills(text):
    words = set(text.split())  # Convert resume text to a set of words
    matched_skills = words.intersection(SKILLS_DB)
    return list(matched_skills)

# Example usage
skills = extract_skills(resume_text)
print("Skills:", skills)
```

```python
from transformers import pipeline

# Load a model that can handle larger inputs
summarizer = pipeline("summarization", model="facebook/mbart-large-50")
```

```python
def summarize_resume(text):
    # Chunk long text if needed
    chunks = [text[i:i+1024] for i in range(0, len(text), 1024)]
    summary = " ".join([summarizer(chunk, max_length=200, min_length=50, do_sample=False)[0]["summary_text"] for chunk in chunks])
    return summary

# Example usage
summary = summarize_resume(resume_text)
print("Summary:", summary)
```

```python
def parse_resume(pdf_path, use_gpt=False):
    resume_text = extract_text_from_pdf(pdf_path)
    entities = extract_entities(resume_text)
    skills = extract_skills(resume_text)
    

    summary = summarize_resume(resume_text)

    return {
        "Name": entities["name"],
        "Email": entities["email"],
        "Phone": entities["phone"],
        "Skills": skills,
        "Summary": summary
    }

# Example usage
parsed_resume = parse_resume("Tech_6.pdf", use_gpt=True)  # Set `use_gpt=False` to use open-source model
print(parsed_resume)
```

```python
import re
import spacy
import pdfplumber
from collections import defaultdict
from nltk.tokenize import sent_tokenize

# Load NLP Model
nlp = spacy.load("en_core_web_sm")

# Predefined Keywords for Scoring
TECH_SKILLS = {"python", "sql", "machine learning", "deep learning", "tensorflow", "pytorch", "nlp"}
PROJECTS = {"project", "model", "classification", "regression", "analysis"}
MATH_STATS = {"probability", "statistics", "linear algebra", "optimization"}
DATA_SKILLS = {"data cleaning", "data preprocessing", "pandas", "numpy"}
COMMUNICATION = {"presentation", "report", "documentation"}
INTERNSHIPS = {"intern", "research", "experience"}
CODING = {"leetcode", "hackerrank", "kaggle", "competitive programming"}
CERTIFICATIONS = {"coursera", "udemy", "kaggle", "certificate"}
EDUCATION = {"bachelor", "master", "phd", "computer science", "data science"}

# Score Weights (Adjustable)
weights = {
    "technical_skills": 2,
    "projects": 1.5,
    "math_stats": 1.2,
    "data_skills": 1.3,
    "communication": 1,
    "internships": 1.5,
    "coding": 1.2,
    "certifications": 1,
    "education": 1.3,
}

# Function to Extract Text from PDF
def extract_text_from_pdf(pdf_path):
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text() + "\n"
    return text

# Function to Compute Section-wise Scores
def score_resume(text):
    text = text.lower()
    scores = defaultdict(int)
    
    # Sentence Tokenization
    sentences = sent_tokenize(text)
    
    # Check for keywords in each section
    for sentence in sentences:
        for skill in TECH_SKILLS:
            if skill in sentence:
                scores["technical_skills"] += 1
        for project in PROJECTS:
            if project in sentence:
                scores["projects"] += 1
        for math in MATH_STATS:
            if math in sentence:
                scores["math_stats"] += 1
        for data in DATA_SKILLS:
            if data in sentence:
                scores["data_skills"] += 1
        for comm in COMMUNICATION:
            if comm in sentence:
                scores["communication"] += 1
        for intern in INTERNSHIPS:
            if intern in sentence:
                scores["internships"] += 1
        for code in CODING:
            if code in sentence:
                scores["coding"] += 1
        for cert in CERTIFICATIONS:
            if cert in sentence:
                scores["certifications"] += 1
        for edu in EDUCATION:
            if edu in sentence:
                scores["education"] += 1

    # Normalize Scores
    max_possible = max(scores.values()) if scores else 1
    for key in scores:
        scores[key] = (scores[key] / max_possible) * 10  # Convert to scale of 10

    # # Weighted Final Score
    final_score = sum(scores[key] * weights[key] for key in scores) / sum(weights.values())

    return scores, round(final_score, 2)
```

```python
# Usage Example
pdf_path = "Tech_6.pdf"  # Change to your resume file path
resume_text = extract_text_from_pdf(pdf_path)
section_scores, total_score = score_resume(resume_text)

print(f"\n**Final Resume Score: {total_score}/10**")
```
